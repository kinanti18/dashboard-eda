### Explore Data Customers

````python
# Distribution of customers (by zip code, city, state)

# by zip code
zipcode_distribution = customers_df['customer_zip_code_prefix'].value_counts().reset_index()
zipcode_distribution.columns = ['Zip Code', 'Number of Customers']

# by city
city_distribution = customers_df['customer_city'].value_counts().reset_index()
city_distribution.columns = ['City', 'Number of Customers']

# by state
state_distribution = customers_df['customer_state'].value_counts().reset_index()
state_distribution.columns = ['State', 'Number of Customers']

# Print the distributions
print("Geographical Distribution by Zip Code:")
print(zipcode_distribution)

print("\nGeographical Distribution by City:")
print(city_distribution)

print("\nGeographical Distribution by State:")
print(state_distribution)
````

````python
# Save file to csv for geographical distribution of customers (by zip code, city, state)
zipcode_distribution.to_csv("zipcode_distribution.csv", index=False)
city_distribution.to_csv("city_distribution.csv", index=False)
state_distribution.to_csv("state_distribution.csv", index=False)
````

````python
# Top 3 Most Common Customer States
common_states = customers_df['customer_state'].value_counts().head(3)
print(common_states)
````

````python
# save file
common_states.to_csv("top_3_most_common_customer_states.csv", index=False)
````

````python
# Top 3 Most Common Customer Cities
common_cities = customers_df['customer_city'].value_counts().head(3)
print(common_cities)
````

````python
# save file
common_cities.to_csv("top_3_most_common_customer_cities.csv", index=False)
````

````python
# Merge the 'orders_df' and 'payments_df' on 'order_id' to get order payment details
merged_df = pd.merge(orders_df, order_payments_df, on='order_id')

# Calculate the total payment (order value) for each order
merged_df['total_payment'] = merged_df['payment_value']

# Group orders by 'customer_id' and calculate the average order value per customer
average_order_value = merged_df.groupby('customer_id')['total_payment'].mean().reset_index()
average_order_value.columns = ['Customer ID', 'Average Order Value']
````

````python
# save file
average_order_value.to_csv("average_order_value_per_customer.csv", index=False)
````

### Explore Data Products

````python
# Merge the 'order_items_df' and 'products_dataset_df' on 'product_id' to get product details
merged_df = pd.merge(order_items_df, products_df, on='product_id')

# Group products by 'product_category_name' and calculate the total quantity sold in each category
top_selling_categories = merged_df.groupby('product_category_name')['order_item_id'].sum().reset_index()
top_selling_categories.columns = ['Product Category', 'Total Quantity Sold']

# Sort the categories by total quantity sold in descending order to find the top sellers
top_selling_categories = top_selling_categories.sort_values(by='Total Quantity Sold', ascending=False).head(3)
````

````python
print(top_selling_categories)
````

````python
# save file
top_selling_categories.to_csv("top_3_selling_categories.csv", index=False)
````

### Explore Data Order

````python
# Convert the relevant date columns to datetime objects
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])

# Calculate the delivery time for each order in days
orders_df['delivery_time'] = (orders_df['order_delivered_customer_date'] - orders_df['order_purchase_timestamp']).dt.days

# Calculate the average delivery time
average_delivery_time = orders_df['delivery_time'].mean()
````

````python
# Group orders by 'order_id' and calculate the average number of items per order
average_items_per_order = order_items_df.groupby('order_id')['order_item_id'].count().mean()
````

````python
# Group payments by 'payment_type' and count the number of payments for each type
payment_counts = order_payments_df['payment_type'].value_counts().reset_index()
payment_counts.columns = ['Payment Type', 'Number of Payments']
````

````python
# save file
payment_counts.to_csv("payment_counts.csv", index=False)
````

### Explore Data Seller

````python
# Merge the 'order_items_df' and 'order_reviews_df' on 'order_id' to get order item details and review scores
merged_df = pd.merge(order_items_df, order_reviews_df, on='order_id')

# Group sellers by 'seller_id' and calculate total sales and average review score
seller_performance = merged_df.groupby('seller_id').agg({'order_item_id': 'sum', 'review_score': 'mean'}).reset_index()
seller_performance.columns = ['Seller ID', 'Total Sales', 'Average Review Score']

# Sort sellers by total sales in descending order
top_sellers_sales = seller_performance.sort_values(by='Total Sales', ascending=False).head(10)

# Sort sellers by average review score in descending order
top_sellers_satisfaction = seller_performance.sort_values(by='Average Review Score', ascending=False).head(10)
````

````python
# save file
top_sellers_satisfaction.to_csv("top_sellers_satisfaction.csv", index=False)
````

````python
# Group sellers by state and count the number of sellers in each state
seller_distribution = sellers_df['seller_state'].value_counts().reset_index()
seller_distribution.columns = ['State', 'Number of Sellers']
````

````python
# save file
seller_distribution.to_csv("seller_distribution.csv", index=False)
````

### Explore Data Delivery Performance

````python
# Convert date columns to datetime objects
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])

# Calculate whether orders were delivered on or before the estimated delivery date
orders_df['on_time_delivery'] = orders_df['order_delivered_customer_date'] <= orders_df['order_estimated_delivery_date']

# Calculate the percentage of on-time deliveries
on_time_delivery_percentage = (orders_df['on_time_delivery'].sum() / len(orders_df)) * 100
````

### Explore Data Translation

````python
# Merge the 'products_df' and 'translations_df' on 'product_category_name' to get product details and translations
merged_df = pd.merge(products_df, product_category_name_translation_df, on='product_category_name', how='left')

# Calculate how often the English translation is used compared to the original category names
original_count = merged_df['product_category_name'].count()
translation_count = merged_df['product_category_name_english'].count()
percentage_translation_used = (translation_count / original_count) * 100
````
